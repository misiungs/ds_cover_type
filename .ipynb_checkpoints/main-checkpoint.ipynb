### Quick data check

```python
#define colnames during of the data set
colnames=['Elevation', 'Aspect', 'Slope',
          'Horizontal_Distance_To_Hydrology',
         'Vertical_Distance_To_Hydrology',
         'Horizontal_Distance_To_Roadways',
         'Hillshade_9am',
         'Hillshade_Noon',
         'Hillshade_3pm',
         'Horizontal_Distance_To_Fire_Points']
colnames.extend(['Wilderness_Area_{:d}'.format(x) for x in range(1,5)])
colnames.extend(['Soil_Type_{:d}'.format(x) for x in range(1,41)])
colnames.append('Cover_Type')
df = pd.read_csv('covtype.data.gz', compression='gzip', names=colnames,
                      header=None, sep=',', quotechar='"', error_bad_lines=False)
```

Data set normalization with no danger of data leakage.

```python
df['Hillshade_9am'] = df['Hillshade_9am']/255.
df['Hillshade_Noon'] = df['Hillshade_Noon']/255.
df['Hillshade_3pm'] = df['Hillshade_3pm']/255.
```

Time for split

```python
#Split data
from sklearn.preprocessing import OneHotEncoder
ran = 33
X = df.loc[:, df.columns != 'Cover_Type']
y = df.loc[:, df.columns == 'Cover_Type']
#first one hot encode the output
y = OneHotEncoder().fit_transform(y).toarray()
#enc = OneHotEncoder().fit(y)
X_trainB, X_test, y_trainB, y_test = train_test_split(X, y, test_size=10000,
                                shuffle=True, stratify=y, random_state=ran)
#decrease train size for hyperparameter tuning
#we will use 90000 for train and 10000 for validation
X_train, _, y_train, _ = train_test_split(X_trainB, y_trainB, train_size=100000,
                                shuffle=True, stratify=y_trainB, random_state=ran)

#small set for initial hyperparameter tuning
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=10000,
                                shuffle=True, stratify=y_train, random_state=ran)

#big set for final tuning
X_trainB, X_valB, y_trainB, y_valB = train_test_split(X_trainB, y_trainB, test_size=10000,
                                shuffle=True, stratify=y_trainB, random_state=ran)
```

Data scalling with prevention of data leakage.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

scaler = ColumnTransformer([
        ('scaler', StandardScaler(), [
        'Elevation', 'Aspect', 'Slope',
        'Horizontal_Distance_To_Hydrology',
        'Vertical_Distance_To_Hydrology',
        'Horizontal_Distance_To_Roadways',
        'Horizontal_Distance_To_Fire_Points'])
    ], remainder='passthrough')

scalerB = ColumnTransformer([
        ('scaler', StandardScaler(), [
        'Elevation', 'Aspect', 'Slope',
        'Horizontal_Distance_To_Hydrology',
        'Vertical_Distance_To_Hydrology',
        'Horizontal_Distance_To_Roadways',
        'Horizontal_Distance_To_Fire_Points'])
    ], remainder='passthrough')

X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)

X_trainB = scalerB.fit_transform(X_trainB)
X_valB = scalerB.transform(X_valB)
X_testB = scalerB.transform(X_test)
```

 We add macro f1-score because it treats all metrics equally. Moreover, micro f1-score is equal accuracy for multilabel classification.

```python
metrics_list = [
        tf.keras.metrics.CategoricalAccuracy(name='accuracy'),
        tf.keras.metrics.Recall(name='recall'), 
        tf.keras.metrics.Precision(name='precision'),
        tfa.metrics.F1Score(num_classes=7, average='macro', name='f1_metric')
]
verb = 0
```

Let's define a function for convergence investigation.

```python
def plot_conv(hist):
    history_dict = hist.history
    
    acc = history_dict['accuracy']
    val_acc = history_dict['val_accuracy']
    prec = history_dict['precision']
    val_prec = history_dict['val_precision']
    rec = history_dict['recall']
    val_rec = history_dict['val_recall']
    f1m = history_dict['f1_metric']
    val_f1m = history_dict['val_f1_metric']
    loss = history_dict['loss']
    val_loss = history_dict['val_loss']
                           
    fig, axs = plt.subplots(2,2, figsize=(15,10))
    epochs = range(1, len(acc) + 1)

    axs[0][0].plot(epochs, val_loss, 'b', label='Validation loss')
    axs[0][0].plot(epochs, loss, c='orange', label='Training loss')
    axs[0][0].set_title('Training and validation loss')
    axs[0][0].set_xlabel('Epochs')
    axs[0][0].set_ylabel('Loss')
    axs[0][0].set_xscale('log')
    axs[0][0].legend()
    
    axs[0][1].plot(epochs, val_acc, 'b', label='Validation accuracy')
    axs[0][1].plot(epochs, acc, c='orange', label='Training accuracy')
    axs[0][1].set_title('Training and validation accuracy')
    axs[0][1].set_xlabel('Epochs')
    axs[0][1].set_ylabel('Loss')
    axs[0][1].set_xscale('log')
    axs[0][1].legend()

    axs[1][0].plot(epochs, val_prec, 'b', label='Validation precision')
    axs[1][0].plot(epochs, prec, c='orange', label='Training precision')
    axs[1][0].plot(epochs, val_rec, c='green', label='Validation recall')
    axs[1][0].plot(epochs, rec, c='violet', label='Training recall') 
    axs[1][0].set_title('Training and validation precision & recall')
    axs[1][0].set_xlabel('Epochs')
    axs[1][0].set_ylabel('Loss')
    axs[1][0].set_xscale('log')
    axs[1][0].legend()
    
    axs[1][1].plot(epochs, val_f1m, 'b', label='Validation f1_score')
    axs[1][1].plot(epochs, f1m, c='orange', label='Training f1_score')
    axs[1][1].set_title('Training and validation f1_score')
    axs[1][1].set_xlabel('Epochs')
    axs[1][1].set_ylabel('Loss')
    axs[1][1].set_xscale('log')
    axs[1][1].legend()
    
    plt.tight_layout()

    print("Final statistics")
    print("Training")
    print("Loss: {:.3f}  Accuracy: {:.3f}  Precision: {:.3f}  Recall: {:.3f}  F1-score:\
  {:.3f}".format(loss[-1], acc[-1], prec[-1], rec[-1], f1m[-1]))
    print("Validation")
    print("Loss: {:.3f}  Accuracy: {:.3f}  Precision: {:.3f}  Recall: {:.3f}  F1-score:\
  {:.3f}".format(val_loss[-1], val_acc[-1], val_prec[-1], val_rec[-1], val_f1m[-1]))
```

### Baseline model.

```python
start_time = time.time()

base_model = tf.keras.Sequential([
  layers.InputLayer(54),
  layers.Dense(50,activation='elu'),
  layers.Dense(7, activation='softmax')
])

base_model.compile(
    optimizer=tf.keras.optimizers.Adam(0.01),
    loss='categorical_crossentropy',
    metrics=metrics_list
)


base_hist = base_model.fit(
    X_train,
    y_train,
    epochs=500,
    batch_size = 128,
    validation_data=(X_val, y_val),
    verbose = 0
)

plot_conv(base_hist)

duration = time.time() - start_time
print("Training time is: {:d} sec".format(int(duration)))
```

```python
#exemplary decaying rates
def decayed_learning_rate(step):
  return initial_learning_rate * decay_rate**(step / decay_steps)

initial_learning_rate = 0.01
decay_rate = 0.1
decay_steps= 1000

step = np.linspace(0,1400)
plt.figure(figsize = (8,6))
plt.plot(step, decayed_learning_rate(step))

initial_learning_rate = 0.01
decay_rate = 0.1
decay_steps= 1400

plt.plot(step, decayed_learning_rate(step))
#plt.yscale('log')
plt.ylim([0,max(plt.ylim())])
plt.xlabel('Epoch')
_ = plt.ylabel('Learning Rate')
```

```python
start_time = time.time()

initial_learning_rate = 0.01
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate,
    decay_steps=500*1400, #number of steps in a batch
    decay_rate=0.1,
    staircase=False)

base_model = tf.keras.Sequential([
  layers.InputLayer(54),
  layers.Dense(50,activation='elu'),
  layers.Dense(40,activation='elu'),
  layers.Dense(7, activation='softmax')
])

base_model.compile(
    optimizer=tf.keras.optimizers.Adam(lr_schedule),
    loss='categorical_crossentropy',
    metrics=metrics_list
)


base_hist = base_model.fit(
    X_train,
    y_train,
    epochs=1400,
    batch_size = 128,
    validation_data=(X_val, y_val),
    verbose = verb
)

plot_conv(base_hist)

duration = time.time() - start_time
print("Training time is: {:d} sec".format(int(duration)))
```

### L2 regularization
